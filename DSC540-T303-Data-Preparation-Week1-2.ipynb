```python
## 3. Create a line chart with Matplotlib and the following data file.
## a. Data file: world-population.xlsm

# Import required libraries.
# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python
# pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
# built on top of the Python programming language

import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Read data from Excel file
file_path = "world-population.xlsm" 
data = pd.read_excel(file_path)

# Step 2: Extract columns (columns are 'Year' and 'Population')
years = data["Year"]
population = data["Population"]

# Step 3: Create a line chart
plt.figure(figsize=(10, 5))  # Set figure size
plt.plot(years, population, marker="o", linestyle="-", color="blue", label="World Population")

# Step 4: Customize the chart
plt.title("World Population Growth", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()  # Adjust layout to prevent clipping

# Step 5: Display the chart
plt.show()
```

```python
# Repeat the experiment 10 times, see the output each time.
# define contant for number of experiments.
NUMBER_OF_EXPERIMENTS = 10

# Declare empty lit to store the different for each experiments
difference_list = []
for i in range(0, NUMBER_OF_EXPERIMENTS):
    random_number_list = [random.randint(0, 100) for x in range(0, 100)]
    list_with_divisible_by_3 = [a for a in random_number_list if a % 3 == 0]
    
    length_of_random_list = len(random_number_list)
    length_of_3_divisible_list = len(list_with_divisible_by_3)
    difference = length_of_random_list - length_of_3_divisible_list
    difference_list.append(difference)

# Print the data from the list
difference_list
```

```python
# Calculate the average difference
avg_diff = sum(difference_list) / float(len(difference_list))
avg_diff
```

```python
# import zip_longest from itertools
from itertools import zip_longest  
```

```python

# create a function 'return_dict_from_csv_line' which will take 2 inputs -> 1)  header values 2) line 
def return_dict_from_csv_line(header, line):
    # Zip them. The zip_longest function is part of the itertools module in Python. 
    # It takes multiple iterables as input and returns an iterator that aggregates elements from each iterable. 
    # The key difference between zip_longest and the standard zip function is how they handle iterables of different lengths.
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict  
```

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # Here we loop over the first 10 lines in order to not to make the output too big
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
